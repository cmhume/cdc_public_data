# Using SODAPy and SODA API to access CDC Public COVID-19 Dataset


Data Source Link:

https://data.cdc.gov/Case-Surveillance/COVID-19-Case-Surveillance-Public-Use-Data/vbim-akqf/data


Tutorials Used:

https://dev.socrata.com/foundry/data.cdc.gov/vbim-akqf


SODA API url: "https://data.cdc.gov/resource/vbim-akqf.json"

```python
# Install sodapy.
!pip install sodapy
```

```python
# Import dependencies.
import pandas as pd
from sodapy import Socrata
```

```python
# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cdc.gov", None)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cdc.gov,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("vbim-akqf", limit=2000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```
